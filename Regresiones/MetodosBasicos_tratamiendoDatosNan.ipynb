# **Preprocesamiento de datos**

Identificacion de campos nulos en todo el dataframe

```python

dataset.isnull().sum() 
```

```python

dataset['Seats'].isnull().sum() # identificando la cantidad de datos nuelos de la columna especifica
```

```python
#reemplazando por la media de los valores
media_seats= dataset['Seats'].mean() #hayando la media de los valores de la columna
# viendo la media
media_seats
```

```python
#reemplazando nan por la amedia
dataset['Seats'].replace(np.nan, media_seats, inplace= True)
```
